```python
# STEP 2: Import Required Libraries
import nltk
import re
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
```

```python
# Download NLTK Data
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# STEP 3: LOAD TEXT FILE
file_name = r"C:\Users\VAIBHAV ADMANE\Downloads\doc1.txt"   # ✅ Change path if needed
with open(file_name, "r") as f:
    text = f.read()
```

```python
print("==============================================")
print("📄 ORIGINAL TEXT")
print("==============================================")
print(text)
```

```python
# STEP 4: TOKENIZATION (Breaking text into words)
print("\n==============================================")
print("🔸 TOKENIZATION (Breaking text into words, 2 words per line)")
print("==============================================")

# Convert to lowercase
text_lower = text.lower()

# Tokenize text
tokens = nltk.word_tokenize(text_lower)

# Print 2 tokens per line
n = 2
for i in range(0, len(tokens), n):
    print(" ".join(tokens[i:i+n]))
```

```python
# STEP 5: REMOVE PUNCTUATION
print("\n==============================================")
print("🔸 REMOVE PUNCTUATION")
print("==============================================")

# Keep only alphabetic tokens (remove punctuation/numbers)
tokens_no_punc = [word for word in tokens if word.isalpha()]

# Print 2 tokens per line
for i in range(0, len(tokens_no_punc), n):
    print(" ".join(tokens_no_punc[i:i+n]))
```

```python
# STEP 6: REMOVE STOPWORDS
print("\n==============================================")
print("🔸 REMOVE STOPWORDS")
print("==============================================")

stop_words = set(stopwords.words('english'))
tokens_no_stop = [word for word in tokens_no_punc if word not in stop_words]

# Print 2 tokens per line
for i in range(0, len(tokens_no_stop), n):
    print(" ".join(tokens_no_stop[i:i+n]))

print("\nTotal Tokens after stopword removal:", len(tokens_no_stop))
```

```python
# STEP 7: STEMMING (Reduce words to root form)
print("\n==============================================")
print("🔸 STEMMING (Reducing words to root form)")
print("==============================================")

ps = PorterStemmer()
tokens_stemmed = [ps.stem(word) for word in tokens_no_stop]

# Print 2 tokens per line
for i in range(0, len(tokens_stemmed), n):
    print(" ".join(tokens_stemmed[i:i+n]))

print("\nTotal Tokens after stemming:", len(tokens_stemmed))
```

```python
# STEP 8: WORD FREQUENCY ANALYSIS
print("\n==============================================")
print("📊 WORD FREQUENCY ANALYSIS")
print("==============================================")

word_freq = Counter(tokens_stemmed)

# Display top 10 frequent words
for word, freq in word_freq.most_common(10):
    print(f"{word}: {freq}")
```

```python
# STEP 9: VISUALIZATION - BAR GRAPH OF TOP 15 WORDS
print("\n==============================================")
print("📊 BAR GRAPH OF TOP 15 WORDS")
print("==============================================")

top_words = word_freq.most_common(15)
words = [w for w, f in top_words]
freqs = [f for w, f in top_words]

plt.figure(figsize=(10,6))
plt.bar(words, freqs, color='skyblue', edgecolor='black')
plt.title("Top 15 Words Frequency After Preprocessing", fontsize=14)
plt.xlabel("Words", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# STEP 10: VISUALIZATION - WORD CLOUD
print("\n==============================================")
print("☁️ WORD CLOUD OF PREPROCESSED TEXT")
print("==============================================")

# Join stemmed words back into a single string
wordcloud_text = " ".join(tokens_stemmed)

# Generate and display the word cloud
wordcloud = WordCloud(width=900, height=500, background_color='white').generate(wordcloud_text)
plt.figure(figsize=(10,7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of Preprocessed Text", fontsize=20)
plt.show()
```
